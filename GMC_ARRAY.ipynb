```python
'''Write a NumPy program to compute the sum of the diagonal element of a given array.'''

import numpy as np
import random

My_Array = np.random.randint(1,7, size = (4,4))
print(My_Array)
My_Diag = My_Array[[0,1,2,3], [0,1,2,3]]
print(My_Diag)
print(np.sum(My_Diag))

```

```python
'''Given two arrays A&B having the same shape. The task is to apply addition of A and B : C is the new array. '''

import numpy as np
import random

A = np.random.randint(1,7, size = (4,4))
B = np.random.randint(7, 30, size = (4,4))
print('This is the First Array A: ', A)
print('This is the First Array B: ', B)
C = A+B
print('This is the result of the summation of A and B: ', C)
```
