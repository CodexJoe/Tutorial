```python
'''Q1: Write a Python function to find the Max of three numbers.'''
'''Using Lambda method'''

def Maxim(num):
    Max = num[0]
    for i in range(1, len(num)):
        if num[i] > Max:
            Max = num[i]
       
    return Max

print(Maxim([7,17,9]))
    
        #OR

def Maxi(*num):# *num made it easy to accept the range of numbers in tuples
    max = 0
    for i in num:
        if i > max:
            max = i
    return max

print(Maxi(7,17,9))
```

```python
'''
Q2: Write a function calculation() such that it can accept two variables and calculate 
    the addition and subtraction of it. And also it must return both addition and subtraction 
in a single return call. For example: calculation(40, 10) should produce 50, 30'''

def calculation(a, b):
    Addition = a + b
    Substraction = a - b
    return Addition, Substraction

calculation(40, 10)
```

```python
'''
Q3:  Write a function that sums the elements of a list of integers.

* Write a function that multiplies the elements of an integer list.

* Use the two functions to sum the elements whose position is an even number (0,2,4â€¦) and multiply the rest.'''

def Summation(num):
    Sum = 0
    for i in num:
        Sum = Sum + i
    return Sum

print(Summation([3,5,6,8,15,2,18]))

def Multiply(num):
    Prod = 1
    for i in num:
        Prod *= i
    return Prod

print(Multiply([3,5,6,8,15,2,18]))

num = [3,5,6,8,15,2,18]

Newlist = list(filter(lambda x: x%2 == 0, num)) # produces another list of elements whose position is an even number of the provious list

print(Summation(Newlist))

Newlist2 = list(filter(lambda x: x%2 != 0, num)) # produces another list of elements whose position is an odd number of the provious list

print(Multiply(Newlist2))
```

```python
'''Q4: Write a Python program that accepts a hyphen-separated sequence of words as input and 
prints the words in a hyphen-separated sequence after sorting them alphabetically.

Sample Items : green-red-yellow-black-white

Expected Result : black-green-red-white-yellow'''

def Sort_Col(colors):
    color_Hy_split = colors.split('-')
    color_Hy_split.sort()
    print(color_Hy_split)#Use this to test progress of the coding
    result = '-' .join(color_Hy_split)
    return result

Get = input("Please enter colors separated by hyphen: ")
Sort_Col(Get)
    
    
```
