```python
import pandas as pd

female = pd.read_csv('daily-total-female-births.csv', header=0, parse_dates=[0], index_col=0)
female.tail()
```

```python
print(female.size)
```

```python
series = female
print(series['1959-01'])
```

```python
series.describe()
```

```python
from matplotlib import pyplot
pyplot.figure(figsize=(10,5))
pyplot.plot(series, color = 'tab:orange') # creates a time series plot with the number of daily births on the y-axis and time in days along the x-axis.
pyplot.gca().set(title='Monthly births in 1959', xlabel='Date', ylabel='Births')
pyplot.show()
```

```python
import matplotlib.pyplot as plt
def plot_timeseries(df, x, y, title="", xlabel='Date', ylabel='Births', dpi=100):
    plt.figure(figsize=(10,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_timeseries(series, x=series.index, y=series.Births, title='Monthly births in 1959')
```
